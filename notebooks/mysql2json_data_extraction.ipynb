### Generate random cut with docstring

```python
import random
```

```python
#json_file="/workspaces/chat-gpt-failures/datasets/semeru/galeras/code_rationals/code_completion_with_docstring_5k_[30_512_tokens].json"
df= pd.read_json(json_file)
```

```python
def generate_random_subcode(code):
    splits = code.split('\n')
    start = len(splits[0]) + 1
    if len(splits) > 2:
        start = len(splits[0]) + random.randint(0,len(splits[1])) + 1 ## Cut after the second line
    #print(code)
    end = random.randint(start, len(code) - 1) ## If you get an error here please check the test bed
    return code[:end]
```

```python
df['random_cut'] = df['code'].apply(generate_random_subcode)
```

```python
df.head()
```

```python
df['prompt'] =  'Generate Pyhton code that ' + df['docstring'] +  ' and code starts with ' + df['random_cut'] +':'
```

```python
#json_file="/workspaces/chat-gpt-failures/datasets/semeru/galeras/code_rationals/code_completion_with_docstring_random_cut_5k_[30_512_tokens].json"
df.to_json(json_file, orient='records')
```
