```python
from Levenshtein import distance as lev
import pandas as pd

```

```python
import json
json_path = '/workspaces/chat-gpt-failures/datasets/galeras_prompting/random_cut.json'
with open(json_path) as json_file:
    json_data = json.load(json_file)
```

```python
lev_calc = [lev(x["code"].strip(), x["predicted_control"] ) for x in json_data]
```

```python
json_path = '/workspaces/chat-gpt-failures/datasets/galeras_prompting/random_cut_P2.json'
with open(json_path) as json_file:
    json_data = json.load(json_file)
```

```python
lev_calc_T1 = [lev(x["code"].strip(), x["predicted_P2"] ) for x in json_data]
```

```python
for i,object in enumerate(json_data):
    object["Levenshtein_control"] = lev_calc[i]
    object["Levenshtein_T1"] = lev_calc_T1[i]
```
